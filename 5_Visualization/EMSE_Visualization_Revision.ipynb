```python
import math
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig


plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare data: extract layers, neuron indices, and scores
layers = list(neuron_data.keys())
n_layers = len(layers)
n_neurons = max(len(vals) for vals in neuron_data.values())

# Initialize arrays for scores and annotations
score_matrix = np.full((n_layers, n_neurons), np.nan)
annot_matrix = [['' for _ in range(n_neurons)] for _ in range(n_layers)]

for i, layer in enumerate(layers):
    data = neuron_data[layer]
    for j, (neuron_idx, score) in enumerate(data):
        score_matrix[i, j] = score
        # Format annotation: neuron index on top, score below
        annot_matrix[i][j] = f"{neuron_idx}\n{score:.2f}"

# Create the heatmap
fig, ax = plt.subplots(figsize=(10, 6))
# Use a perceptually-uniform colormap
cax = ax.imshow(score_matrix, cmap='viridis', aspect='auto', vmin=0.23, vmax=1.0)

# Set ticks: columns as rank (1 to 10), rows as layer names
ax.set_xticks(np.arange(n_neurons))
ax.set_xticklabels([f"Rank {i+1}" for i in range(n_neurons)], fontsize=9, rotation=45)
ax.set_yticks(np.arange(n_layers))
ax.set_yticklabels(layers, fontsize=10)

# Annotate each cell with neuron index and score
for i in range(n_layers):
    for j in range(n_neurons):
        if not np.isnan(score_matrix[i, j]):
            ax.text(j, i, annot_matrix[i][j],
                    ha='center', va='center', color='w', fontsize=8, fontweight='bold')

# Add a colorbar and label
cbar = fig.colorbar(cax, ax=ax, fraction=0.046, pad=0.04)
cbar.set_label('Suspiciousness Score', fontsize=10)

ax.set_title('Top-10 Neuron-Level Suspiciousness Scores by Layer', fontsize=12)
plt.tight_layout()
plt.show()
```
