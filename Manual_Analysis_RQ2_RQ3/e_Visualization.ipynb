```python
import math

from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig


plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

```python
DLSW_MRR = [37.1, 31.6, 42.3, 45.5]
NDLSW_MRR = [52.6, 45.3, 61.1, 61.7]

index = ["BugLocator", "BLUIR", "BLIA", "DNNLOC"]

# Creating figure and axis objects
fig, ax = plt.subplots(figsize=(10,6))

# Defining width of each bar and their position
width = 0.2
x = np.arange(len(index))
DLSW_colors = ['darkslategrey']
NDLSW_colors = ['teal']
# Plotting the bars

ax.bar(x-width, DLSW_MRR, width, label='DLSW_MRR',color=DLSW_colors)
ax.bar(x, NDLSW_MRR, width, label='NDLSW_MRR',color=NDLSW_colors)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Setting the axis labels and title
ax.set_xlabel('Method')
ax.set_ylabel('MRR (%)')

# Setting the x-axis tick labels
ax.set_xticks(x)
ax.set_xticklabels(index)

DLSW_patch = plt.Rectangle((0,0),1,1,fc=DLSW_colors[0], edgecolor = 'none')
NDLSW_patch = plt.Rectangle((0,0),1,1,fc=NDLSW_colors[0], edgecolor = 'none')


ax.legend([DLSW_patch, NDLSW_patch], ['DLSW', 'NDLSW'])
plt.show()
```
